# 2. Vendas por Categoria


```python
'''
Calcular a quantidade de produtos vendidos por categoria em cada loja.
A ideia é agrupar os dados por categoria e contar o número de vendas
de cada tipo, mostrando as categoria mais populares de cada loja
'''
import csv
from collections import Counter
from urllib.request import urlopen
import io

# Função para contar categorias de um arquivo csv acessado por URL
def contar_categorias(url):
    categorias = []
    with urlopen(url) as resposta:
        texto = io.TextIOWrapper(resposta, encoding='utf-8')
        arquivo_csv = csv.DictReader(texto)
        for linha in arquivo_csv:
            categorias.append(linha["Categoria do Produto"])
    return Counter(categorias)

# Dicionário com os nomes das lojas e suas URLs
urls_lojas = {
    "Loja 1": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_2.csv",
    "Loja 2": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_2.csv",
    "Loja 3": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_3.csv",
    "Loja 4": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_4.csv",
}

# Conta e exibe os resultados
for loja, url in urls_lojas.items():
    print(f"{loja} - Produtos vendidos por categoria:")
    contagem = contar_categorias(url)
    for categoria, qtd in contagem.items():
        print(f"  {categoria}: {qtd} vendas")
    print()
```

# 3. Média de Avaliação das Lojas

```python
'''
Calcular a média das avaliações dos clientes para cada loja.
O objetivo é entender a satisfação dos clientes com os produtos vendidos.
'''
import csv
from urllib.request import urlopen
import io

# Função para calcular a média das avaliações a partir de uma URL
def media_avaliacoes_compra(url):
    avaliacoes = []

    # Abri o arquivo CSV via URL
    with urlopen(url) as resposta:
        texto = io.TextIOWrapper(resposta, encoding='utf-8')
        arquivo_csv = csv.DictReader(texto)

        # Coleta os valores da coluna "Avaliação da compra"
        for linha in arquivo_csv:
            try:
                nota = float(linha["Avaliação da compra"])  # Converte para float
                avaliacoes.append(nota)
            except ValueError:
                pass  # Ignora valores vazios ou inválidos

    # Calcula a média das avaliações de compra
    if len(avaliacoes) == 0:
        return 0
    else:
        return sum(avaliacoes) / len(avaliacoes)

# Dicionário com os nomes das lojas e as URLs dos arquivos
urls_lojas = {
    "Loja 1": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_1.csv",
    "Loja 2": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_2.csv",
    "Loja 3": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_3.csv",
    "Loja 4": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_4.csv",
}

# Mostra a média de cada loja
for loja, url in urls_lojas.items():
    media = media_avaliacoes_compra(url)
    print(f"{loja} - Média de Avaliação das compras: {media:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
'''
Identifica quais produtos foram os mais vendidos e os menos vendidos em cada loja.
Ao final, mostra os resultados dequais produtos se destacaram nas vendas de cada loja.
'''
import csv
from urllib.request import urlopen
import io
from collections import Counter

# Função para identificar os produtos mais e menos vendidos
def produtos_mais_menos_vendidos(url):
    produtos = []

    # Abri e lê o arquivo csv a partir da URL
    with urlopen(url) as resposta:
        texto = io.TextIOWrapper(resposta, encoding='utf-8')
        arquivo_csv = csv.DictReader(texto)
        for linha in arquivo_csv:
            produtos.append(linha["Produto"])

    # Conta quantas vezes cada produto foi vendido
    contagem = Counter(produtos)

    if not contagem:
        return [], [], contagem

    # Maior e menor número de vendas
    max_vendas = max(contagem.values())
    min_vendas = min(contagem.values())

    # Lista dos mais vendidos e menos vendidos
    mais_vendidos = [(produto, qtd) for produto, qtd in contagem.items() if qtd == max_vendas]
    menos_vendidos = [(produto, qtd) for produto, qtd in contagem.items() if qtd == min_vendas]

    return mais_vendidos, menos_vendidos, contagem

# Dicionário com as URLs das lojas
urls_lojas = {
    "Loja 1": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_1.csv",
    "Loja 2": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_2.csv",
    "Loja 3": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_3.csv",
    "Loja 4": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_4.csv",
}

# Executa e exibe os resultados
for loja, url in urls_lojas.items():
    mais, menos, _ = produtos_mais_menos_vendidos(url)

    print(f"\n {loja}")

    print("Produto(s) mais vendido(s):")
    for produto, qtd in mais:
        print(f"   - {produto}: {qtd} vendas")

    print("Produto(s) menos vendido(s):")
    for produto, qtd in menos:
        print(f"   - {produto}: {qtd} venda(s)")
```

# 5. Frete Médio por Loja


```python
'''
Calcular o custo médio de frete para cada loja.
O objetivo é entender quanto, em média, está sendo gasto com frete para cada uma das lojas.
'''
import csv
from urllib.request import urlopen
import io

# Função para calcular o custo médio de frete de uma loja
def calcular_media_frete(url):
    total_frete = 0
    quantidade = 0

    # Abri e le o CSV via URL
    with urlopen(url) as resposta:
        texto = io.TextIOWrapper(resposta, encoding='utf-8')
        arquivo_csv = csv.DictReader(texto)

        # Soma os valores da coluna "Frete"
        for linha in arquivo_csv:
            try:
                frete = float(linha["Frete"])
                total_frete += frete
                quantidade += 1
            except ValueError:
                pass  # Ignora valores vazios ou inválidos

    # Calcula média
    if quantidade == 0:
        return 0
    else:
        return total_frete / quantidade

# Dicionário com as lojas e URLs
urls_lojas = {
    "Loja 1": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_1.csv",
    "Loja 2": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_2.csv",
    "Loja 3": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_3.csv",
    "Loja 4": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_4.csv",
}

# Calcula e mostra o custo médio de frete de cada loja
for loja, url in urls_lojas.items():
    media_frete = calcular_media_frete(url)
    print(f"{loja} - Custo médio do frete: R$ {media_frete:.2f}")
```

```python
#Calcula o faturamento por loja e e custo médio do frete
import csv
from urllib.request import urlopen
import io
import matplotlib.pyplot as plt
from collections import defaultdict, Counter

# URLs das lojas
urls_lojas = {
    "Loja 1": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_1.csv",
    "Loja 2": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_2.csv",
    "Loja 3": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_3.csv",
    "Loja 4": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_4.csv",
}

# Inicializando estruturas
faturamento_por_loja = {}
frete_medio_por_loja = {}

# Coletando os dados
for loja, url in urls_lojas.items():
    total_frete = 0
    total_preco = 0
    total_linhas = 0

    with urlopen(url) as resposta:
        texto = io.TextIOWrapper(resposta, encoding='utf-8')
        leitor = csv.DictReader(texto)

        for linha in leitor:
            try:
                preco = float(linha["Preço"])
                frete = float(linha["Frete"])
                total_preco += preco
                total_frete += frete

            except:
                continue

    faturamento_por_loja[loja] = total_preco
    frete_medio_por_loja[loja] = total_frete / total_linhas if total_linhas > 0 else 0
```

```python
# ========== GRÁFICO 1 - Faturamento por Loja ==========
plt.figure(figsize=(8, 5))
plt.bar(faturamento_por_loja.keys(), faturamento_por_loja.values(), color='#c74634')
plt.title("Faturamento Total por Loja")
plt.xlabel("Lojas")
plt.ylabel("R$ Faturamento")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# ========== GRÁFICO 2 - Custo Médio do Frete ==========
plt.figure(figsize=(8, 5))
plt.plot(list(frete_medio_por_loja.keys()), list(frete_medio_por_loja.values()), marker='o', linestyle='-', color='green')
plt.title("Custo Médio de Frete por Loja")
plt.xlabel("Lojas")
plt.ylabel("R$ Frete Médio")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# ========== GRÁFICO 3 - Distribuição de vendas por Loja ==========
import csv
from urllib.request import urlopen
import io
from collections import Counter
import matplotlib.pyplot as plt

# URLs das lojas
urls_lojas = {
    "Loja 1": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_1.csv",
    "Loja 2": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_2.csv",
    "Loja 3": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_3.csv",
    "Loja 4": "https://github.com/lucaskrlima/challenge1-data-science-aluraStore-one/raw/main/data/raw/loja_4.csv",
}

# Armazena os produtos mais vendidos
mais_vendidos = {}

for loja, url in urls_lojas.items():
    contador = Counter()
    with urlopen(url) as resposta:
        texto = io.TextIOWrapper(resposta, encoding='utf-8')
        arquivo_csv = csv.DictReader(texto)
        for linha in arquivo_csv:
            produto = linha["Produto"]
            contador[produto] += 1
    produto, quantidade = contador.most_common(1)[0]
    mais_vendidos[f"{loja} - {produto}"] = quantidade

# Dados para o gráfico
labels = list(mais_vendidos.keys())
valores = list(mais_vendidos.values())

# Cria gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(valores, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Participação dos Produtos Mais Vendidos por Loja")
plt.axis('equal')  # Para manter o formato redondo
plt.tight_layout()
plt.show()
```
